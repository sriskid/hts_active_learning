```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# 5% Sampling

```python
# Random sampling performance plot
rand_acc = np.load("random_performance10.npy")[:475]
rand_std = np.load("random_std10.npy")[:475]
rand_acc.shape
```

```python
# Reinforcement learning performance plot
rein_acc = np.load("rein_rewa_performance10.npy")[:475]
rein_std = np.load("rein_rewa_std10.npy")[:475]
rein_acc.shape
```

```python
# QBC
qbc_acc = np.load("qbc/QBC_Accuracy_mean.npy")
qbc_std = np.load("qbc/QBC_Accuracy_std.npy")
qbc_acc.shape
```

```python
# Uncertainty
uncer_acc = np.load("uncertainty/uncer_performance.npy")[:475]
uncer_std = np.load("uncertainty/uncer_std.npy")[:475]
uncer_acc.shape
```

```python
N_rounds = 475
Initial_train_size = 26
xaxis = np.arange(Initial_train_size, N_rounds+Initial_train_size)
xaxis.shape
```

```python
plt.figure()
# print(avg_acc.shape, xaxis.shape)
# Create plot
sns.set_style("whitegrid")
# Plot the data with shaded error bars
# Plot the data with error bars
# Create a mask to show error bars only every 50 data points
# Subset arrays to show error bars every 50 on the x-axis
step_size = 50

x_subset = xaxis[::step_size]
y_subset = uncer_acc[::step_size]
std_subset = uncer_std[::step_size]

sns.lineplot(x=xaxis, y=uncer_acc, errorbar='ci', err_style='bars',color='b', label='Uncertainty Sampling')
plt.errorbar(x=x_subset, y=y_subset, yerr=std_subset, fmt='none', color='b', capsize=5)


x_subset = xaxis[::step_size]
y_subset = rand_acc[::step_size]
std_subset = rand_std[::step_size]

sns.lineplot(x=xaxis, y=rand_acc, errorbar='ci', err_style='bars', color='black', label='Random Learning')
plt.errorbar(x=x_subset, y=y_subset, yerr=std_subset, fmt='none', color='black', capsize=5)

x_subset = xaxis[::step_size]
y_subset = rein_acc[::step_size]
std_subset = rein_std[::step_size]

sns.lineplot(x=xaxis, y=rein_acc, errorbar='ci', err_style='bars',color='r', label='Reinforcement Active Learning')
plt.errorbar(x=x_subset, y=y_subset, yerr=std_subset, fmt='none', color='r', capsize=5)

x_subset = xaxis[::step_size]
y_subset = qbc_acc[::step_size]
std_subset = qbc_std[::step_size]

sns.lineplot(x=xaxis, y=qbc_acc, errorbar='ci', err_style='bars',color='c', label='Query-by-committee')
plt.errorbar(x=x_subset, y=y_subset, yerr=std_subset, fmt='none', color='c', capsize=5)




# adjust legend location
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.ylim([0.4, 1])
plt.ylabel("Accuracy")
plt.xlabel("No. of samples in training set")
plt.title("Accuracy")

plt.savefig("Accuracy_plot")
```
